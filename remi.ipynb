```python
def tokenized(text):
    tok = word_tokenize(text)
    s = []
    
    for j in tok:
        if j not in stop_words and j in valid_tokens:
            s.append(j)
            
    return s



def compare_phrase(user_text,texts):
    similarity = []
    tok_user_text = tokenized(user_text)
    for i in range(len(texts)):
        v = model.n_similarity(texts[i],tok_user_text)
        if v == 1:
            pass
        else:
            similarity.append((v,data_questions["Id"][i]))
    return max(similarity)
```
